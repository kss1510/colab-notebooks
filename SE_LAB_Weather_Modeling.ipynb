```python
# Weather modeling using quadratic temperature model
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-darkgrid')  # Nice background

def quadratic_temperature_model(day, a, b, c):
    return a * day**2 + b * day + c

def main():
    # Fixed coefficients
    a, b, c = 1, -5, 6  # ax^2 + bx + c

    # Ask user for number of days
    num_days = int(input("Enter the number of days to model: "))

    # Generate day and temperature data
    days = list(range(1, num_days + 1))
    temperatures = [quadratic_temperature_model(day, a, b, c) for day in days]

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(days, temperatures, 'o-', color='crimson', linewidth=2, markersize=6,
             label=f'Temperature = {a}x + ({b})x + {c}')

    plt.title('Weather Forecast Modeling (Quadratic) ', fontsize=16, fontweight='bold')
    plt.xlabel('Day', fontsize=12)
    plt.ylabel(' Temperature (C)', fontsize=12)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

main()
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-darkgrid')  # Set a nice background style

# Quadratic temperature model: ax^2 + bx + c
def quadratic_temperature_model(day, a, b, c):
    return a * day**2 + b * day + c

def main():
    print("Quadratic Temperature Model: ax + bx + c")

    # Taking user input for coefficients
    a = float(input("Enter coefficient a: "))
    b = float(input("Enter coefficient b: "))
    c = float(input("Enter coefficient c: "))

    # Number of days
    num_days = int(input("Enter number of days to model: "))

    # Generate list of days
    days = list(range(1, num_days + 1))
    temperatures = [quadratic_temperature_model(day, a, b, c) for day in days]

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(days, temperatures, 'o-', color='orange', linewidth=2, markersize=6,
             label=f'Temperature = {a}x + {b}x + {c}')

    plt.title('Weather Forecast Modeling', fontsize=16, fontweight='bold')
    plt.xlabel(' Day', fontsize=12)
    plt.ylabel(' Temperature (C)', fontsize=12)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

main()
```

```python
# Weather modeling from CSV with single set of coefficients
import matplotlib.pyplot as plt
import pandas as pd

# Set a beautiful plot style
plt.style.use('fivethirtyeight')

def quadratic_weather_model(time, a, b, c):
    return a * time**2 + b * time + c

def main():
    print("Quadratic Weather Modeling (From File)")
    print("=========================================")

    try:
        file_path = 'wea2.csv'  # Make sure this file is uploaded in Colab
        df = pd.read_csv(file_path)

        time_values = list(range(0, 11))  # Time from 0 to 10
        plt.figure(figsize=(10, 6))

        # Assuming the CSV has one row with columns: a, b, c
        for index, row in df.iterrows():
            a, b, c = row['a'], row['b'], row['c']
            temperatures = [quadratic_weather_model(t, a, b, c) for t in time_values]

            plt.plot(time_values, temperatures, marker='o', linestyle='-', linewidth=2, color='darkblue',
                     label=f'Temperature = {a}x + {b}x + {c}')
            break  # Read only the first row (single set of inputs)

        # Enhancing plot
        plt.title('Temperature Variation Over Time', fontsize=16, fontweight='bold')
        plt.xlabel('Time (days)', fontsize=12)
        plt.ylabel('Temperature (C)', fontsize=12)
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.xlim(0, 10)
        plt.legend()
        plt.tight_layout()
        plt.show()

    except FileNotFoundError:
        print("File not found. Please upload 'wea2.csv' in the Colab environment.")

main()
```

```python
# Weather modeling using multiple quadratic equations from a CSV
import matplotlib.pyplot as plt
import pandas as pd

# Use visually appealing style
plt.style.use('ggplot')

def quadratic_weather_model(time, a, b, c):
    return a * time**2 + b * time + c

def main():
    print(" Quadratic Weather Modeling (Multiple Models)")
    print("===============================================")

    try:
        file_path = 'wea.csv'  # Make sure this file is uploaded to Colab
        df = pd.read_csv(file_path)

        time_values = list(range(0, 11))  # 0 to 10 days
        plt.figure(figsize=(10, 6))

        for index, row in df.iterrows():
            a, b, c = row['a'], row['b'], row['c']
            temperature_values = [quadratic_weather_model(t, a, b, c) for t in time_values]

            # Distinct color and label for each set
            plt.plot(
                time_values,
                temperature_values,
                marker='o',
                linestyle='--',
                linewidth=2,
                label=f'{index+1}) {a}x + {b}x + {c}'
            )

        # Enhancing the plot
        plt.title('Temperature Variation Over Time (Multiple Models)', fontsize=16, fontweight='bold')
        plt.xlabel('Time (days)', fontsize=12)
        plt.ylabel('Temperature (C)', fontsize=12)
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.xlim(0, 10)
        plt.legend()
        plt.tight_layout()
        plt.show()

    except FileNotFoundError:
        print(" File not found. Please upload 'wea.csv' into the Colab environment.")

main()
```
